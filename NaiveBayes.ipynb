```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

# import the tweet dataset
X1 = pd.read_csv("D:\\360DigiTMG\\DataScience\\17. Machine Learning Classifier Technique - Naive Bayes\\Assignment Q\\Disaster_tweets_NB.csv")
X1.head()
```

```python
# dropping the unusual columns
X1.drop(['keyword', 'location', 'id'], axis=1, inplace=True)
X1.head()
```

```python
# text preprocessing

"""
Here extracting the stopwords from a file

then splitted that and stored as an array
"""
stop_words = []
with open("D:\\360DigiTMG\\DataScience\\0. Personal\\stopwords.txt") as sw:
    stop_words = sw.read()

stop_words = stop_words.split('\n')
stop_words
```

```python
# custom function to clean the tweets
import re
def cleanText(i):
    i = re.sub("[^A-Za-z" "]+", " ", i).lower()
    i = re.sub("[0-9" "]+"," ",i)
    words = []
    for w in i.split(" "):
        if w not in stop_words:
            words.append(w)
    return " ".join(words)

cleanText("Hi my name is Ritendu Bhattacharyya")
```

```python
X1.text = X1.text.apply(cleanText)
X1.head()
```

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test = train_test_split(X1, test_size=0.2)
```

```python
# custom function to split into words 
def split_words(w):
    return w.split(" ")

X1_bow = CountVectorizer(analyzer=split_words).fit(X1.text)

X1_all_bow_matrix = X1_bow.transform(X1.text)
X1_train_bow_matrix = X1_bow.transform(X1_train.text)
X1_test_bow_matrix = X1_bow.transform(X1_test.text)

tfidf = TfidfTransformer().fit(X1_all_bow_matrix)
tfidf_train = tfidf.transform(X1_train_bow_matrix)
tfidf_test = tfidf.transform(X1_test_bow_matrix)
```

```python
# creating navie_bayes model
from sklearn.naive_bayes import MultinomialNB as MB

mb = MB()
mb.fit(tfidf_train, X1_train.target)

test_predict_mb = mb.predict(tfidf_test)
test_predict_mb
```

```python
# accuracy checks
from sklearn.metrics import accuracy_score as score
score(X1_test.target, test_predict_mb)
```

```python
pd.crosstab(test_predict_mb, X1_test.target)
```

```python
train_predict_mb = mb.predict(tfidf_train)
train_predict_mb
```

```python
# accuracy checks
from sklearn.metrics import accuracy_score as score
score(X1_train.target, train_predict_mb)
```

```python
pd.crosstab(train_predict_mb, X1_train.target)
```

```python
# creating navie_bayes model with tunning parameter
from sklearn.naive_bayes import MultinomialNB as MB

mb = MB(alpha=2)
mb.fit(tfidf_train, X1_train.target)

test_predict_mb = mb.predict(tfidf_test)
test_predict_mb
```

```python
score(X1_test.target, test_predict_mb)
```
